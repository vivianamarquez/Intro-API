# 5. `POST` Requests in Python

A `POST` request sends data to an API.

We usually use `POST` when we want to create something, submit a form, or send a message to a server.


```python
import json
import requests
from pprint import pprint

TIMEOUT = 10
```

## Send JSON Data

JSONPlaceholder lets us practice `POST` requests without an account. It simulates creating a post and returns the data we sent plus a new `id`.


```python
url = "https://jsonplaceholder.typicode.com/posts"
payload = {
    "title": "Learning APIs",
    "body": "POST requests send data to an API.",
    "userId": 7,
}

response = requests.post(url, json=payload, timeout=TIMEOUT)

print(response.status_code)
print(response.headers["Content-Type"])
```

```python
created_post = response.json()
pprint(created_post)
```

A common successful status code for `POST` is `201`, which means something was created.


```python
print(response.status_code == 201)
print(created_post["id"])
```

## What Did Python Send?

When we use the `json=` argument, `requests` converts our dictionary to JSON and sets the content type header for us.


```python
print(response.request.method)
print(response.request.headers["Content-Type"])
print(response.request.body.decode())
```

## Echo a Request Back

httpbin can send our request data back to us. This is useful for learning because we can inspect exactly what arrived at the API.


```python
echo_url = "https://httpbin.org/post"
echo_payload = {
    "student_name": "Ada",
    "topic": "APIs",
    "ready_for_post": True,
}

echo_response = requests.post(echo_url, json=echo_payload, timeout=TIMEOUT)
echo_response.raise_for_status()

echo_data = echo_response.json()
pprint(echo_data["json"])
```

## Send Form Data

Some APIs expect form data instead of JSON. Use `data=` for that style of request.


```python
form_payload = {
    "first_name": "Ada",
    "favorite_method": "POST",
}

form_response = requests.post(echo_url, data=form_payload, timeout=TIMEOUT)
form_response.raise_for_status()

form_data = form_response.json()
pprint(form_data["form"])
```

## Put `POST` Logic in a Function

The function below keeps the payload and status check in one place.


```python
def create_demo_post(title, body, user_id):
    url = "https://jsonplaceholder.typicode.com/posts"
    payload = {
        "title": title,
        "body": body,
        "userId": user_id,
    }
    response = requests.post(url, json=payload, timeout=TIMEOUT)

    if response.status_code == 201:
        return response.json()

    print(f"Expected 201, but got {response.status_code}.")
    return None

new_post = create_demo_post(
    title="Class notes",
    body="Today we practiced POST requests.",
    user_id=3,
)

pprint(new_post)
```

## Important Note

JSONPlaceholder is a practice API. It responds like it created the post, but it does not permanently save new data.

That makes it useful for class: students can practice sending real HTTP requests without needing logins or changing a production database.
