# 4. `GET` Requests in Python

A `GET` request asks an API for data.

We usually use `GET` when we want to read information without changing anything on the server.


```python
import requests
from pprint import pprint

TIMEOUT = 10
```

## Basic `GET`

This request asks JSONPlaceholder for one sample post.


```python
url = "https://jsonplaceholder.typicode.com/posts/1"
response = requests.get(url, timeout=TIMEOUT)

print(response.status_code)
print(response.headers["Content-Type"])
```

```python
post = response.json()
pprint(post)
```

```python
print(post["title"])
print(post["body"])
```

## `GET` With Query Parameters

Query parameters let us filter or customize what the API sends back.


```python
comments_url = "https://jsonplaceholder.typicode.com/comments"
params = {"postId": 1}

comments_response = requests.get(comments_url, params=params, timeout=TIMEOUT)
comments = comments_response.json()

print(comments_response.status_code)
print(comments_response.url)
print(len(comments))
```

```python
for comment in comments[:3]:
    print(comment["email"])
    print(comment["body"][:80])
    print("---")
```

## A Real-World Public Data API

This request asks for public holidays in the United States for 2025.


```python
holidays_url = "https://date.nager.at/api/v3/PublicHolidays/2025/US"
holidays_response = requests.get(holidays_url, timeout=TIMEOUT)
holidays_response.raise_for_status()

holidays = holidays_response.json()
print(len(holidays))
```

```python
for holiday in holidays[:6]:
    print(f"{holiday['date']} - {holiday['name']}")
```

## Filter the Data With Python

After the API returns data, we can use Python to answer questions.


```python
july_holidays = []

for holiday in holidays:
    if holiday["date"].startswith("2025-07"):
        july_holidays.append(holiday)

pprint(july_holidays)
```

## Put `GET` Logic in a Function

Functions help us reuse the same request with different values.


```python
def get_comments_for_post(post_id):
    url = "https://jsonplaceholder.typicode.com/comments"
    params = {"postId": post_id}
    response = requests.get(url, params=params, timeout=TIMEOUT)
    response.raise_for_status()
    return response.json()

comments_for_post_2 = get_comments_for_post(2)
print(len(comments_for_post_2))
print(comments_for_post_2[0]["email"])
```

## Quick Review

For a `GET` request, we usually care about:

- the URL
- optional query parameters
- the status code
- the JSON response
